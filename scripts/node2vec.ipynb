```python
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import networkx as nx
import pickle
import os
from node2vec import Node2Vec
from sklearn.metrics import jaccard_score
import seaborn as sns
from gensim.models import KeyedVectors
from matplotlib_venn import venn3
```

```python
input_folder = "yearly_graphs_new"
output_folder = "node2vec_results"
os.makedirs(output_folder, exist_ok=True)

# read in the research dictionary
female_words = pd.read_csv('female_words.csv')
female_words = female_words.sort_values(by='total_count', ascending=False).head(100)  # select top 100 for node2vec
female_tokens = set(female_words['token'].tolist())

# read in the frequency data
token_year_counts = pd.read_csv('token_year_counts.csv')

yearly_similar_sets = {}

# iterate over all yearly graphs
file_list = [f for f in os.listdir(input_folder) if f.startswith("graph_") and f.endswith(".pkl")]
for file_name in tqdm(file_list, desc="Processing graphs"):
    year = int(file_name.split('_')[1].split('.')[0])
    if year < 1949:
        continue

    file_path = os.path.join(input_folder, file_name)

    # read in the yearly graph
    with open(file_path, 'rb') as f:
        yearly_graph = pickle.load(f)

    # get the top 1000 tokens for the year
    top_tokens = token_year_counts[token_year_counts['year'] == year]
    top_tokens = top_tokens.sort_values(by='count', ascending=False).head(1000)['token'].tolist()

    # merge the two sets and keep only the unique tokens
    nodes_to_keep = set(female_tokens) | set(top_tokens)
    subgraph = yearly_graph.subgraph(nodes_to_keep).copy()

    # train the yearly Node2Vec models
    node2vec = Node2Vec(subgraph, dimensions=64, walk_length=30, num_walks=200, workers=4)
    model = node2vec.fit(window=10, min_count=1, batch_words=4)

    # save the models
    model_save_path = os.path.join(output_folder, f"node2vec_{year}.model")
    model.wv.save(model_save_path)

    # find out the most similar nodes to the female nodes
    similar_sets = set()
    for female_node in female_tokens:
        if female_node in model.wv:
            similar_nodes = model.wv.most_similar(female_node, topn=50)  # get the top 50 most similar nodes
            similar_sets.update([node for node, _ in similar_nodes])

    yearly_similar_sets[year] = similar_sets

# Calculate the Jaccard Similarity for different years
years = sorted(yearly_similar_sets.keys())
num_years = len(years)
jaccard_matrix = pd.DataFrame(index=years, columns=years, dtype=float)

for i in tqdm(range(num_years), desc="Calculating Jaccard Similarity"):
    for j in range(num_years):
        year1, year2 = years[i], years[j]
        set1, set2 = yearly_similar_sets[year1], yearly_similar_sets[year2]
        if set1 and set2:
            intersection = len(set1 & set2)
            union = len(set1 | set2)
            jaccard_similarity = intersection / union if union > 0 else 0
        else:
            jaccard_similarity = 0
        jaccard_matrix.loc[year1, year2] = jaccard_similarity

# visualize the Jaccard Similarity across years
plt.figure(figsize=(12, 10))
sns.heatmap(jaccard_matrix, cmap="Blues", annot=False, cbar=True)
plt.title("Jaccard Similarity Between Years", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Year", fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
yearly_similar_sets = {}
similarity_long_data = []

# exclude women related words from the graph before calculating the similarity
exclude_pattern = "妇|母|娘|女|妈|婆|奶|姥|姐|妹|姑|姨|嫂|婶|妻|巾帼|太太"

file_list = [f for f in os.listdir(output_folder) if f.startswith("node2vec_") and f.endswith(".model")]
for file_name in tqdm(file_list, desc="Loading models"):
    year = int(file_name.split('_')[1].split('.')[0])

    # load the models
    model_path = os.path.join(output_folder, file_name)
    model = KeyedVectors.load(model_path)

    avg_similarity = {}
    for node in model.key_to_index:
        # exclude the related words
        if pd.Series([node]).str.contains(exclude_pattern).any():
            continue
        similarities = [model.similarity(node, female_node) for female_node in female_tokens if female_node in model]
        if similarities:
            avg_similarity[node] = sum(similarities) / len(similarities)

    sorted_avg_similarity = sorted(avg_similarity.items(), key=lambda x: x[1], reverse=True)
    similar_sets = {node for node, similarity in sorted_avg_similarity[:1000]}

    for node, similarity in sorted_avg_similarity[:1000]:
        similarity_long_data.append({
            "Year": year,
            "Node": node,
            "Average Similarity": similarity
        })

    yearly_similar_sets[year] = similar_sets

# save data in csv files
similarity_long_df = pd.DataFrame(similarity_long_data)
similarity_long_df.to_csv("yearly_similarity_long.csv", index=False)

years = sorted(yearly_similar_sets.keys())
num_years = len(years)
jaccard_matrix = pd.DataFrame(index=years, columns=years, dtype=float)

for i in tqdm(range(num_years), desc="Calculating Jaccard Similarity"):
    for j in range(num_years):
        year1, year2 = years[i], years[j]
        set1, set2 = yearly_similar_sets[year1], yearly_similar_sets[year2]
        if set1 and set2:
            intersection = len(set1 & set2)
            union = len(set1 | set2)
            jaccard_similarity = intersection / union if union > 0 else 0
        else:
            jaccard_similarity = 0
        jaccard_matrix.loc[year1, year2] = jaccard_similarity

plt.figure(figsize=(12, 10))
sns.heatmap(jaccard_matrix, cmap="Blues", annot=False, cbar=True)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Year", fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# plot the historical events of Cultural Revolution and Reform and Opening Up
plt.axvline(x=years.index(1966), color='red', linestyle='--')
plt.axvline(x=years.index(1977), color='red', linestyle='--')
plt.axhline(y=years.index(1966), color='red', linestyle='--')
plt.axhline(y=years.index(1977), color='red', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
# read in the similarity data and tokens with POS
similarity_long_df = pd.read_csv("yearly_similarity_long.csv")
tokens_with_pos = pd.read_csv("tokens_with_pos.csv")

# filter out the nouns
adjective_tokens = tokens_with_pos[tokens_with_pos["pos"].str.contains("n")]["token"]

# define the three time periods
periods = {
    "Before 1966": range(1949, 1966),
    "1966-1976": range(1966, 1977),
    "After 1976": range(1977, similarity_long_df["Year"].max() + 1)
}

# get the nouns subset for each period
period_sets = {}
for period_name, years in periods.items():
    period_words = similarity_long_df[similarity_long_df["Year"].isin(years)]["Node"]
    period_adjectives = set(period_words[period_words.isin(adjective_tokens)])
    period_sets[period_name] = period_adjectives

# calcualte the intersection of the sets
before_1966 = period_sets["Before 1966"]
during_1966_1976 = period_sets["1966-1976"]
after_1976 = period_sets["After 1976"]

only_before_1966 = before_1966 - during_1966_1976 - after_1976
only_during_1966_1976 = during_1966_1976 - before_1966 - after_1976
only_after_1976 = after_1976 - before_1966 - during_1966_1976
shared_between_before_and_during = (before_1966 & during_1966_1976) - after_1976
shared_between_during_and_after = (during_1966_1976 & after_1976) - before_1966
shared_between_before_and_after = (before_1966 & after_1976) - during_1966_1976
shared_all_three = before_1966 & during_1966_1976 & after_1976

# select representative words for each category
representative_words = {
    "Only Before 1966": list(only_before_1966)[:10],
    "Only 1966-1976": list(only_during_1966_1976)[:10],
    "Only After 1976": list(only_after_1976)[:10],
    "Before & During": list(shared_between_before_and_during)[:10],
    "During & After": list(shared_between_during_and_after)[:10],
    "Before & After": list(shared_between_before_and_after)[:10],
    "All Three Periods": list(shared_all_three)[:10],
}

# plot the Venn diagram and fill in the representative words
plt.figure(figsize=(10, 10))
venn = venn3(
    subsets=(
        len(only_before_1966),
        len(only_during_1966_1976),
        len(shared_between_before_and_during),
        len(only_after_1976),
        len(shared_between_before_and_after),
        len(shared_between_during_and_after),
        len(shared_all_three),
    ),
    set_labels=("Before 1966", "1966-1976", "After 1976"),
)

venn.get_label_by_id("100").set_text("\n".join(representative_words["Only Before 1966"]))
venn.get_label_by_id("010").set_text("\n".join(representative_words["Only 1966-1976"]))
venn.get_label_by_id("001").set_text("\n".join(representative_words["Only After 1976"]))
venn.get_label_by_id("110").set_text("\n".join(representative_words["Before & During"]))
venn.get_label_by_id("011").set_text("\n".join(representative_words["During & After"]))
venn.get_label_by_id("101").set_text("\n".join(representative_words["Before & After"]))
venn.get_label_by_id("111").set_text("\n".join(representative_words["All Three Periods"]))

plt.show()
```
